```python
def train_valid_split(dframe, seed):
    train_data = pd.DataFrame()
    valid_data = pd.DataFrame()
    np.random.seed(seed)
    for i in tqdm(range(len(dframe)-1)):
        idx1 = np.random.randint(len(dframe) - 1)
        idx2 = idx1 +1
        
        row1 = dframe.iloc[[idx1]].reset_index()
        row2 = dframe.iloc[[idx2]].reset_index()
        
        randInt = np.random.randint(9)
        if 0 <= randInt <= 1:
            valid_frames = [valid_data, row1, row2]
            valid_data = pd.concat(valid_frames, axis = 0, join = 'outer', ignore_index=False)
        if randInt >= 2:
            train_frames = [train_data, row1, row2]
            train_data = pd.concat(train_frames, axis = 0, join = 'outer', ignore_index=False)
    return train_data, valid_data
```

### Transformations

```python
def transform_brightness(img, brightness):
    hsv_image = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
    hsv_image[:,:,2] = hsv_image[:,:,2] * brightness
    rgb_image = cv2.cvtColor(hsv_image, cv2.COLOR_HSV2RGB)
    return rgb_image
```

```python
def optical_flow_dense(img, next_img):
    img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    next_img_gray = cv2.cvtColor(next_img, cv2.COLOR_RGB2GRAY)
    
    # Optical Flow Parameters
    flow_mat = None
    pyr_scale = 0.5
    levels = 1
    winsize = 15
    iterations = 2
    poly_n = 5
    poly_sigma = 1.1
    
    flow = cv2.calcOpticalFlowFarneback(img_gray, next_img_gray, flow_mat, pyr_scale, levels, winsize, iterations, poly_n, poly_sigma, 0)
    
    mag, ang = cv2.cartToPolar(flow[..., 0], flow[..., 1]) 
    
    hsv = np.zeros((66, 220, 3))
    # set saturation
    hsv[:,:,1] = cv2.cvtColor(next_img, cv2.COLOR_RGB2HSV)[:,:,1]
    # hue corresponds to direction
    hsv[:,:,0] = ang * (180/ np.pi / 2)
    # value corresponds to magnitude
    hsv[:,:,2] = cv2.normalize(mag,None,0,255,cv2.NORM_MINMAX)
    
    hsv = np.asarray(hsv, dtype= np.float32)
    rgb_flow = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
    
    return rgb_flow
```

```python
def resize(img):
    img_cropped = img[25:375, :]
    image = cv2.resize(img_cropped, (220, 66), interpolation = cv2.INTER_AREA)
    
    return image
```

```python
def preprocess_train_data(img_path, speed, brightness):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = resize(img)
    img = transform_brightness(img, brightness)
    return img, speed

def preprocess_valid_data(img_path, speed):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = resize(img)
    return img, speed
```

```python
class Dataset(data.Dataset):
    def __init__(self, data, preprocess_train=None, preprocess_valid=None, process_optical_flow=None, valid=False):
        self.data = data
        self.preprocess_train = preprocess_train
        self.preprocess_valid = preprocess_valid
        self.process_optical_flow = process_optical_flow
        self.valid = valid

    def __getitem__(self, index):            
        try: 
            row_now = self.data.iloc[[index]].reset_index()
            row_prev = self.data.iloc[[index-1]].reset_index()
            row_next = self.data.iloc[[index+1]].reset_index()
        except Exception as e:
            print(e)
            
        
        time_now = row_now['image_index'].values[0]
        time_prev = row_prev['image_index'].values[0]
        time_next = row_next['image_index'].values[0]
        
        if abs(time_now - time_prev) == 1 and time_now > time_prev:
            row1 = row_prev
            row2 = row_now

        elif abs(time_next - time_now) == 1 and time_next > time_now:
            row1 = row_now
            row2 = row_next
        else:
            print('Error generating row') 
        
        if self.valid:
            # preprocess first image
            x1, y1 = self.preprocess_valid(row1['image_path'].values[0], row1['speed'].values[0])
            # preprocess second image
            x2, y2 = self.preprocess_valid(row2['image_path'].values[0], row2['speed'].values[0])
        else:
            # brightness factor 
            brightness = 0.2 + np.random.uniform()
            # preprocess first image
            x1, y1 = self.preprocess_train(row1['image_path'].values[0], row1['speed'].values[0], brightness)
            # preprocess second image
            x2, y2 = self.preprocess_train(row2['image_path'].values[0], row2['speed'].values[0], brightness)
        
        # compute optical flow send in images as RGB
        rgb_diff = self.process_optical_flow(x1, x2)
        if self.valid:
            rgb_diff = rgb_diff.reshape(1, rgb_diff.shape[0], rgb_diff.shape[1], rgb_diff.shape[2])

        # calculate mean speed
        y = np.mean([y1, y2])
        
        rgb_diff = (rgb_diff/127.5) - 1
        
        return rgb_diff, y
        
    def __len__(self):
        return len(self.data)
```

```python
params = {'batch_size': 32,
          'shuffle': True,
          'num_workers': 0}

training_set = Dataset(train_data, preprocess_train=preprocess_train_data, process_optical_flow=optical_flow_dense)
training_generator = data.DataLoader(training_set, **params)

validation_set = Dataset(valid_data, preprocess_valid=preprocess_valid_data, process_optical_flow=optical_flow_dense, valid=True)
validation_generator = data.DataLoader(validation_set, **params)
```

```python
def generate_training_data(data, batch_size = 32):
    image_batch = np.zeros((batch_size, 66, 220, 3)) # nvidia input params
    label_batch = np.zeros((batch_size))
    while True:
        for i in range(batch_size):
            # generate a random index with a uniform random distribution from 1 to len - 1
            idx = np.random.randint(1, len(data) - 1)
            
            row_now = data.iloc[[idx]].reset_index()
            row_prev = data.iloc[[idx-1]].reset_index()
            row_next = data.iloc[[idx+1]].reset_index()
            
            time_now = row_now['image_index'].values[0]
            time_prev = row_prev['image_index'].values[0]
            time_next = row_next['image_index'].values[0]

            if abs(time_now - time_prev) == 1 and time_now > time_prev:
                row1 = row_prev
                row2 = row_now

            elif abs(time_next - time_now) == 1 and time_next > time_now:
                row1 = row_now
                row2 = row_next
            else:
                print('Error generating row') 

            # brightness factor 
            brightness = 0.2 + np.random.uniform()
            # preprocess first image
            x1, y1 = preprocess_train_data(row1['image_path'].values[0], row1['speed'].values[0], brightness)
            # preprocess second image
            x2, y2 = preprocess_train_data(row2['image_path'].values[0], row2['speed'].values[0], brightness)

            # compute optical flow send in images as RGB
            rgb_diff = optical_flow_dense(x1, x2)

            # calculate mean speed
            y = np.mean([y1, y2])

            rgb_diff = (rgb_diff/127.5) - 1
            
            image_batch[i] = rgb_diff
            label_batch[i] = y
        
        #print('image_batch', image_batch.shape, ' label_batch', label_batch)
        # Shuffle the pairs before they get fed into the network
        yield shuffle(image_batch, label_batch)
```

```python
def generate_validation_data(data):
    while True:
        for idx in range(1, len(data) - 1): # start from the second row because we may try to grab it and need its prev to be in bounds
            row_now = data.iloc[[idx]].reset_index()
            row_prev = data.iloc[[idx-1]].reset_index()
            row_next = data.iloc[[idx+1]].reset_index()

            time_now = row_now['image_index'].values[0]
            time_prev = row_prev['image_index'].values[0]
            time_next = row_next['image_index'].values[0]

            if abs(time_now - time_prev) == 1 and time_now > time_prev:
                row1 = row_prev
                row2 = row_now

            elif abs(time_next - time_now) == 1 and time_next > time_now:
                row1 = row_now
                row2 = row_next
            else:
                print('Error generating row') 

            # preprocess first image
            x1, y1 = preprocess_valid_data(row1['image_path'].values[0], row1['speed'].values[0])
            # preprocess second image
            x2, y2 = preprocess_valid_data(row2['image_path'].values[0], row2['speed'].values[0])

            # compute optical flow send in images as RGB
            rgb_diff = optical_flow_dense(x1, x2)
            rgb_diff = rgb_diff.reshape(1, rgb_diff.shape[0], rgb_diff.shape[1], rgb_diff.shape[2])
            rgb_diff = (rgb_diff/127.5) - 1
            
            # calculate mean speed
            y = np.mean([y1, y2])
            
            speed = np.array([[y]])
            
            #print('img_diff', img_diff.shape, ' speed', speed)
            yield img_diff, speed
```

```python
training_generator = generate_training_data(train_data)
validation_generator = generate_validation_data(valid_data)
```

### NVIDIA Network

```python
class Network(nn.Module):
    def __init__(self):
        super(Network, self).__init__()
        
        self.conv_layers = nn.Sequential(
            nn.Conv2d(3, 24, 5, stride=2),
            nn.ELU(),
            nn.Conv2d(24, 36, 5, stride=2),
            nn.ELU(),
            nn.Conv2d(36, 48, 5, stride=2),
            nn.ELU(),
            nn.Dropout(0.5),
            nn.Conv2d(48, 64, 3),
            nn.ELU(),
            nn.Conv2d(64, 64, 3)
        )
        
        self.linear_layers = nn.Sequential(
            nn.Linear(64*1*20, 100),
            nn.ELU(),
            nn.Linear(100, 50),
            nn.ELU(),
            nn.Linear(50, 10),
            nn.ELU(),
            nn.Linear(10, 1)
        )
    
    def forward(self, input):
        input = input.view(input.size(0), 3, 66, 220)
        output = self.conv_layers(input)
        output = output.view(output.size(0), -1)
        output = self.linear_layers(output)
        return output
        
```

```python
model = Network()
optimizer = optim.Adam(model.parameters(), lr=1e-4)

criterion = nn.MSELoss()
```

```python
max_epochs = 16

for epoch in tqdm(range(max_epochs)):
    
    # Training
    train_loss = 0
    model.train()
    
    for i, (img_batch, speed) in enumerate(training_generator):
        optimizer.zero_grad()
        outputs = model(torch.Tensor(img_batch))
        loss = criterion(outputs, torch.Tensor(speed).unsqueeze(1))
        loss.backward()
        optimizer.step()
        
        train_loss += loss.data.item()
    
        if i % 100 == 0:
            print("Loss: %.3f" % (train_loss/((i+1))))
        

    model.eval()
    valid_loss = 0
    with torch.set_grad_enabled(False):
        for i, (img_diff, speed) in enumerate(validation_generator):
            optimizer.zero_grad()
            outputs = model(img_diff)
            loss = criterion(outputs, speed.unsqueeze(1))
            
            train_loss += loss.data.item()
    
            if i % 100 == 0:
                print("Valid Loss: %.3f" % (valid_loss/((i+1))))
    
```
